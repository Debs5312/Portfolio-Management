# Limits of Diversification

```python
%load_ext autoreload
%autoreload 2
import Edhec_risk_kit as erk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ind_return = erk.industry_dframe()
ind_size = erk.industry_dframe_MarketSize()
ind_nfirms = erk.industry_dframe_firmsNo()
```

```python
ind_return.head(2)
```

```python
ind_size.head(2)
```

```python
ind_nfirms.head(2)
```

```python
ind_mktcap = ind_size*ind_nfirms
ind_mktcap.head()
```

```python
total_mktcap = ind_mktcap.sum(axis=1)
```

```python
total_mktcap.plot(figsize=(15,6))
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis="rows")
```

```python
ind_capweight.head()
```

```python
ind_capweight["1926"].sum(axis="columns")
```

```python
ind_capweight[["Fin","Steel"]].plot(figsize=(15,6))
```

```python
total_market_return = (ind_capweight*ind_return).sum(axis="columns")
total_market_return.head()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(15,6))
```

```python
total_market_index["1986":].plot(figsize=(12,6))
total_market_index["1986":].rolling(window=36).mean().plot(figsize=(12,6))#moving average for 36 months aka 3 years 
```

```python
trailing_36mnths_rets = total_market_return.rolling(window=36).aggregate(erk.periodic_returns, period=12)
trailing_36mnths_rets.plot(figsize=(12,6),label="Trailing 36 months Compound Annualized Return",legend=True)
```

### Correlation VS Market Up-Down Realtion

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ["date","industry"]
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level="date").apply(lambda corrmat: corrmat.values.mean())
ind_tr36corr.plot()
```

```python
trailing_36mnths_rets.plot(figsize=(12,6),label="Trailing 36 months Compound Annualized Return",legend=True)
ind_tr36corr.plot(label="Trailling 36 months Correlation",legend=True, secondary_y = True)
```

#### the Above Pic Clearly Shows that when market is Going Down Correlation Goes up(i.e All Industries are doing bad) and Vice versa... and this is the reason of faliure of Diversification...becaz if we are making A diversified portfolio it will ultimately return loss.
